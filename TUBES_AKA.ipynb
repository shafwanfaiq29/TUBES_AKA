```python
import pandas as pd

file_path = '/Customer-Churn-Records.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:  # Sorting in ascending order
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Menemukan titik tengah
        left_half = arr[:mid]  # Membagi array menjadi dua bagian
        right_half = arr[mid:]

        merge_sort(left_half)  # Panggilan rekursif untuk bagian kiri
        merge_sort(right_half)  # Panggilan rekursif untuk bagian kanan

        i = j = k = 0

        # Menggabungkan dua bagian yang sudah terurut
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Menambahkan sisa elemen dari left_half jika ada
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        # Menambahkan sisa elemen dari right_half jika ada
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

    return arr
```

```python
import time
import random
import matplotlib.pyplot as plt

# Mengambil kolom 'Balance' dari dataset
balances = df['Balance'].dropna().tolist()

# Daftar ukuran dataset yang akan diuji
dataset_sizes = [100, 200, 300, 400, 500]
bubble_sort_times = []
merge_sort_times = []

# Mengukur waktu eksekusi untuk setiap ukuran dataset
for size in dataset_sizes:
    data = random.sample(balances, size)  # Mengambil sampel acak dari data 'Balance'

    # Salinan data untuk Bubble Sort
    data_copy = data.copy()

    # Mengukur waktu untuk Bubble Sort
    start_time = time.time()
    bubble_sort(data_copy)
    bubble_sort_times.append(time.time() - start_time)

    # Salinan data untuk Merge Sort
    data_copy = data.copy()

    # Mengukur waktu untuk Merge Sort
    start_time = time.time()
    merge_sort(data_copy)
    merge_sort_times.append(time.time() - start_time)

# Plot grafik perbandingan waktu eksekusi
plt.plot(dataset_sizes, bubble_sort_times, label="Bubble Sort")
plt.plot(dataset_sizes, merge_sort_times, label="Merge Sort")
plt.xlabel("Ukuran Dataset")
plt.ylabel("Waktu Eksekusi (detik)")


plt.title("Perbandingan Waktu Eksekusi Algoritma Sorting")
plt.legend()
plt.show()
```
